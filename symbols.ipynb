```python
df_symbol = pd.read_csv('data/yahoo_stocklist.csv') #'data/tickers.csv'
display(df_symbol.head(10))
print(len(df_symbol))
```

```python
iwant = (df_symbol['COMPANY NAME'].notnull())
iwant &= (df_symbol['COUNTRY TRADED']=='United States of America')
df_symbol_us = df_symbol[iwant]
print(len(df_symbol_us))
```

```python
def trans_name(name):
    return name.lower().split(',')[0].split(' inc.')[0].split('.com')[0]

def check_name(name1, name2):
    len1, len2 = len(name1), len(name2)
    if len1 == len2:
        return name1 == name2
    elif len1 < len2:
        return (name1 == name2[:len1]) & (not name2[len1].isalnum()) # ensure the word stoped at the same position
    else:
        return False#(name1[:len2] == name2) & (not name1[len2].isalnum())

def find_symbol(name):
    symbols = []
    name1 = trans_name(name)
    if len(symbols) == 0:
        iwant = df_symbol_us['COMPANY NAME'].apply(lambda x: name1 == trans_name(x))
        symbols = df_symbol_us['YAHOO TICKER'][iwant].values
    if len(symbols) == 0:
        iwant = df_symbol_us['YAHOO TICKER'].apply(lambda x: name1 == x.lower())
        symbols = df_symbol_us['YAHOO TICKER'][iwant].values
    if len(symbols) == 0:
        iwant = df_symbol_us['COMPANY NAME'].apply(lambda x: check_name(name1, trans_name(x)) )
        symbols = df_symbol_us['YAHOO TICKER'][iwant].values
    
    symbols = list(set(symbols))
    return symbols
```

```python
df['symbol'] = df['company_name'].apply(lambda name: find_symbol(name))
```
